```python
import pandas as pd
import json

# Load JSON file
file_path = "data4.json"
with open(file_path, "r") as file:
    data = json.load(file)

# Initialize a list to hold expanded data for each document
documents_data = []

# Loop over each document in the 'results' list
for result in data.get("results", []):
    # Prepare dictionary with all relevant fields
    document_info = {
        "document_title": result.get("title"),
        "document_number": result.get("document_number"),
        "action": result.get("action"),
        "publication_date": result.get("publication_date"),
        "pdf_url": result.get("pdf_url"),
        "abstract": result.get("abstract"),
        "excerpts": result.get("excerpts"),
        "topics": ", ".join(result.get("topics", [])),  # Join topics list as a string
        "dates": result.get("dates"),
        "page_length": result.get("page_length"),
        "start_page": result.get("start_page"),
        "end_page": result.get("end_page"),
        "significant": result.get("significant"),
    }

    # Extract and join agencies into a single string
    if "agencies" in result and isinstance(result["agencies"], list):
        document_info["agency_names"] = ", ".join(
            agency.get("name", "") for agency in result["agencies"]
        )
        # Extract additional agency details and join them into strings
        document_info["agency_ids"] = ", ".join(
            str(agency.get("id", "")) for agency in result["agencies"]
        )
        document_info["agency_urls"] = ", ".join(
            agency.get("url", "") for agency in result["agencies"]
        )
        document_info["agency_parent_ids"] = ", ".join(
            str(agency.get("parent_id", "")) for agency in result["agencies"]
        )
        document_info["agency_slugs"] = ", ".join(
            agency.get("slug", "") for agency in result["agencies"]
        )

    # Add CFR references if they exist
    if "cfr_references" in result and isinstance(result["cfr_references"], list):
        document_info["cfr_titles"] = ", ".join(str(ref.get("title", "")) for ref in result["cfr_references"])
        document_info["cfr_parts"] = ", ".join(str(ref.get("part", "")) for ref in result["cfr_references"])

    # Append this enriched data to the main list
    documents_data.append(document_info)

# Convert to DataFrame for easy viewing and manipulation
all_data_df = pd.DataFrame(documents_data)

# Handle missing values by filling them with empty strings (for NLP tasks)
all_data_df.fillna("", inplace=True)

# Display the first few rows of the DataFrame
print(all_data_df.head())

# Optionally, export the DataFrame to a CSV file
output_csv_path = "processed_documents_with_agencies.csv"
all_data_df.to_csv(output_csv_path, index=False)
print(f"Data has been exported to {output_csv_path}")
```

```python
import pandas as pd

# List of file paths
file_paths = [
    "1.csv",
    "2.csv",
    "3.csv",
    "4.csv",
    "5.csv"
]

# Initialize an empty list to hold DataFrames
dataframes = []

# Loop through each file and read it into a DataFrame
for file in file_paths:
    df = pd.read_csv(file)
    dataframes.append(df)

# Concatenate all DataFrames row-wise
result_df = pd.concat(dataframes, axis=0, ignore_index=True)

# Fill any missing values with empty strings
result_df.fillna("", inplace=True)

# Export the combined DataFrame to a new CSV file
output_file = "combined_data.csv"
result_df.to_csv(output_file, index=False)

print(f"Combined CSV file has been saved to: {output_file}")
```
